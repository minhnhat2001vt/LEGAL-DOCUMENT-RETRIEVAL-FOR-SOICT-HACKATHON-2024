```python
import jsonlines
import os

jsonl_directory = "/kaggle/input/train-data-full"  # Replace with your directory path

def count_json_objects_in_file(file_path):
    count = 0
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            if line.strip():  # Ensure the line is not empty
                count += 1
    return count
num_obj_lst = []
for filename in os.listdir(jsonl_directory):
        # Process only files that end with .jsonl
        if filename.endswith(".jsonl"):
            file_path = os.path.join(jsonl_directory, filename)
            number_of_objects = count_json_objects_in_file(file_path)
            num_obj_lst.append(number_of_objects)
            print(f'The file {file_path} contains {number_of_objects} JSON objects.')
```

```python
import json

# Path to your JSONL file
input_file_path = '/kaggle/working/combined_train_data.jsonl'
output_file_path = '/kaggle/working/train_data.jsonl'

# Open the input and output files
with open(input_file_path, 'r', encoding='utf-8') as infile, \
     open(output_file_path, 'w', encoding='utf-8') as outfile:
    for line in infile:
        # Parse each line as a JSON object
        data = json.loads(line.strip())
        
        # Clean the 'pos' field by removing newline characters
        if 'pos' in data:
            data['pos'] = [sentence.replace('\n', ' ') for sentence in data['pos']]
            # Retain only negatives from index 
            data["neg"] = data["neg"][7:]
        
        # Write the cleaned JSON object to the output file
        json.dump(data, outfile, ensure_ascii=False)
        outfile.write('\n')
```

```python
import json

# Define the shortened prompt to append to each sample
prompt_text = (
    "Cho một câu hỏi pháp lý, chọn đoạn văn phù hợp nhất từ danh sách.Một hoặc nhiều đoạn văn tích cực sẽ chứa thông tin đầy đủ để trả lời câu hỏi; các đoạn văn tiêu cực thì không. Mục tiêu là giúp mô hình phân biệt giữa các đoạn văn tích cực và tiêu cực."
)

# Input and output file paths
input_file_path = "/kaggle/working/combined_train_data.jsonl"  # Replace with your input JSONL file path
output_file_path = "/kaggle/working/combined_train_data_with_prompt.jsonl"  # Output file path

# Read, update, and write the data with the shortened prompt added
with open(input_file_path, 'r', encoding='utf-8') as infile, open(output_file_path, 'w', encoding='utf-8') as outfile:
    for line in infile:
        sample = json.loads(line)
        sample['prompt'] = prompt_text  # Update the prompt
        json.dump(sample, outfile, ensure_ascii=False)
        outfile.write('\n')  # Ensure each sample is on a new line
```
