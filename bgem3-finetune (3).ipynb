```python
import json
import random

# Path to your JSONL file
input_file_path = '/kaggle/working/combined_train_data.jsonl'
output_file_path = '/kaggle/working/train_data.jsonl'

# Load all data first to allow random selection of negatives
all_data = []
with open(input_file_path, 'r', encoding='utf-8') as infile:
    for line in infile:
        all_data.append(json.loads(line.strip()))

# Collect all negatives from the entire dataset
all_negatives = [neg for sample in all_data for neg in sample.get("neg", [])]

# Counter for samples with insufficient negatives
insufficient_negatives_count = 0

# Open the output file for writing
with open(output_file_path, 'w', encoding='utf-8') as outfile:
    for data in all_data:
        # Clean the 'pos' field by removing newline characters
        if 'pos' in data:
            data['pos'] = [sentence.replace('\n', ' ') for sentence in data['pos']]
        
        # Retain only negatives from index 6 onwards
        if 'neg' in data:
            data["neg"] = data["neg"][6:]
        
        # Check if the sample has enough negatives
        if len(data["neg"]) < 1:  # Adjust the threshold as needed
            # Increment the counter for insufficient negatives
            insufficient_negatives_count += 1
            
            # Randomly sample negatives from the global pool to fill the gap
            additional_negatives = random.sample(all_negatives, k=1)  # Choose 1 or more as needed
            data["neg"].extend(additional_negatives)

        # Write the cleaned and augmented JSON object to the output file
        json.dump(data, outfile, ensure_ascii=False)
        outfile.write('\n')

# Print the total number of samples with insufficient negatives
print(f"Number of samples with insufficient negatives: {insufficient_negatives_count}")
```

```python
import json

# Path to your JSONL file
file_path = output_file_path

# Initialize a counter for the number of queries processed
query_count = 0

# Open and read the JSONL file
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        # Parse each line as a JSON object
        data = json.loads(line.strip())
        query = data.get('query', 'No query provided')
        positive = data.get('pos', [])
        negatives = data.get('neg', [])

        # Extract the first three negative sentences
        first_three_negatives = negatives[:3]

        # Print the query and its first three negative sentences
        print(f"Query: {query}")
        print (f"Positive: {positive}")
        for i, sentence in enumerate(first_three_negatives, start=1):
            print(f"Negative {i}: {sentence}")
        print('-' * 40)

        # Increment the query count
        query_count += 1

        # Stop after processing 10 queries
        if query_count >= 10:
            break
```
